```python
import matplotlib
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('house_pricing_dataset.csv')
df.columns
```

```python
new_data=df.head(10)
new_data
```

```python
#The distribution of house prices
fig = plt.figure(figsize=(8,6))
plt.hist(df.price,range= (342000,452500),bins=10)
plt.title('Histogram of distribution of house prices ')

```

From the histogram plotted above the following can be deduced:
1.prices between 340000 and 350000 occurred more than 100 times,prices between 400000 and 410000 also occurred more than 100 times,same thing with prices between 430000 and 440000.And they are the most occurring prices
2.Prices between 420000 to 430000,440000 to 450000 both occurred like 90 times.
3.Prices roughly between 375000 to 385000 occurred 70 times and is the least occurring price 
4.Prices between 355000 and 365000 occurred like 75 times 

```python
#The distribution of average house prices across the top 10 cities

new_data.groupby('city',as_index=False )['price'].mean()
plt.bar(new_data.city,new_data.price)
plt.title('Graph of distribution of average house price across top 10 cities ')
plt.xlabel(' City')
plt.ylabel('Price')

```

From the graph above,we can deduce that 
1.seattle is the most priced city 
2.Redmond is the second most priced city and the gap between seattle and Redmond is huge
3.Maple Valley and North Bend are priced at same rate 
4.The least priced city is Shoreline

```python
#The distribution of maximum bedroom size across the top 10 cities
figsize=(12,8)
new_data.groupby('city',as_index=False )['bedrooms'].max()
plt.bar(new_data.city,new_data.bedrooms)
plt.title('Graph of distribution of bedrooms ')
plt.xlabel(' City')
plt.ylabel('Number of Bedrooms')

plt.show()
```

From the graph of distribution of nedrooms above,we can conclude that:
1.Houses in seattle city as the highest number of bedrooms which is 5
2.Houses in Redmond and Maple Valley have the same number of rooms which is 4
3.Houses in Shoreline,Kent,Bellevue and North Bend have the same number of bedrooms which is 3
