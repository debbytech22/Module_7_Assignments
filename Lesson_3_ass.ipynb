```python
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("company_sales_data.csv")
df
```

```python

#Question 1
x_values =np.arange(len(df.month_number))
plt.bar(x_values +0.2,df.facecream,width=0.4)
plt.bar(x_values -0.2,df.facewash,width =0.4)
plt.xticks(x_values)
plt.xlabel(' Months')
plt.ylabel('Price of Toiletries')
plt.title('Graph of sales of facewash and facecream against months')
plt.legend(['facewash','facecream'])
plt.show()
```

```python
#Question 2
yerror=[800,3900,450,1000,2000,2500,1000,100,2000,-500,-3000,-4000]
plt.bar(df.month_number,df.bathingsoap,yerr=yerror,capsize =10)
plt.xticks(df.month_number)
plt.xlabel(' Months')
plt.ylabel('Bathing Soap')
plt.title('Graph of sales Bathingsoap against months')
plt.show()
```

```python
total_fwash=df['facewash'].sum(axis=0)
total_fcream=df['facecream'].sum(axis=0)
total_tpaste=df['toothpaste'].sum(axis=0)
total_bsoap=df['bathingsoap'].sum(axis=0)
total_shamp=df['shampoo'].sum(axis=0)
total_mostzer=df['moisturizer'].sum(axis=0)
total_fwash=df['facewash'].sum(axis=0)
pie_total=[total_fwash,total_fcream,total_tpaste,total_bsoap,total_shamp,total_mostzer]
toiletries_label =['facewash','facecream','toothpaste','bathingsoap','shampoo','moisturizer']
```

```python
#num =[12,15]
plt.pie(pie_total,labels=toiletries_label,autopct='%0.1f%%')
plt.axis('equal')
plt.show()
```
