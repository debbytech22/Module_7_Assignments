```python
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('anscombe_quartet_data.csv')
print(df)
```

METHOD ONE

```python

plt.subplots_adjust(hspace=0.4)
ax = plt.subplot(2, 2, 1)
plt.scatter(df.x1,df.y1)
plt.title('Graph of x1y1')
p1, p0 = np.polyfit(df.x1, df.y1, deg=1)  # slope, intercept
ax.axline(xy1=(0, p0), slope=p1, color='r', lw=2)

ax = plt.subplot(2, 2, 2)
plt.scatter(df.x2,df.y2)
plt.title('Graph of x2y2')
p1, p0 = np.polyfit(df.x1, df.y1, deg=1)  # slope, intercept
ax.axline(xy1=(0, p0), slope=p1, color='g', lw=2)

ax = plt.subplot(2, 2, 3)
plt.scatter(df.x3,df.y3)
plt.title('Graph of x3y3')
p1, p0 = np.polyfit(df.x1, df.y1, deg=1)  # slope, intercept
ax.axline(xy1=(0, p0), slope=p1, color='y', lw=2)

ax = plt.subplot(2, 2, 4)
plt.scatter(df.x4,df.y4)
plt.title('Graph of x4y4')
p1, p0 = np.polyfit(df.x1, df.y1, deg=1)  # slope, intercept
ax.axline(xy1=(0, p0), slope=p1, color='b', lw=2)

plt.show()
```

```python


fig,((ax1,ax2,),(ax3,ax4)) =  plt.subplots(nrows=2, ncols=2)
plt.title('Graph of x1y1')
ax1.scatter(df.x1,df.y1)
p1, p0 = np.polyfit(df.x1, df.y1, deg=1) 
ax1.axline(xy1=(0, p0), slope=p1, color='r', lw=2)



ax2.scatter(df.x2,df.y2)
plt.title('Graph of x2y2')
p1, p0 = np.polyfit(df.x2, df.y2, deg=1) 
ax2.axline(xy1=(0, p0), slope=p1, color='b', lw=2)

ax3.scatter(df.x3,df.y3)
plt.title('Graph of x3y3')
p1, p0 = np.polyfit(df.x3, df.y3, deg=1) 
ax3.axline(xy1=(0, p0), slope=p1, color='g', lw=2)


ax4.scatter(df.x4,df.y4)
plt.title('Graph of x4y4')
p1, p0 = np.polyfit(df.x4, df.y4, deg=1) 
ax4.axline(xy1=(0, p0), slope=p1, color='y', lw=2)
plt.subplots_adjust(hspace=0.9,wspace=0.9)

```
