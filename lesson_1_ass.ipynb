```python
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#reading csv file
df = pd.read_csv("company_sales_data.csv")
print(df)
```

```python
plt.plot( df.month_number,df.facecream,linestyle =':',marker ='o',color ='purple')
plt.plot( df.month_number,df.facewash,linestyle='--',marker ='s',color ='green')
plt.plot( df.month_number,df.toothpaste,linestyle =':',marker ='*',color ='yellow')
plt.plot( df.month_number,df.bathingsoap,linestyle ='--',marker ='s',color ='blue')
plt.plot( df.month_number,df.shampoo,linestyle =':',marker = '*',color ='red')
plt.plot( df.month_number,df.moisturizer,linestyle ='--',marker ='o',color ='black')
plt.xlabel('month_number ')
plt.ylabel('product')
plt.title(' graph of company sales per month')
plt.legend(['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'],loc ='upper left')
plt.show()
```
